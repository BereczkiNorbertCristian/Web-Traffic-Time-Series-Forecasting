```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import matplotlib.patches as mpatches
from fbprophet import Prophet
```

```python
def readFile(filename):
    lst = []
    cnt = 0
    with open(filename,"r") as f:
        for line in f:
            if cnt == 100000: break
            x = float(line.rstrip("\n"))
            if x > 10: continue
            if math.isnan(x): continue
            lst.append(x)
            cnt+=1
    return lst
```

```python
many_results = readFile("OUT/partial.csv")
results_pd = pd.DataFrame(many_results)

plt.figure(figsize=(20,10))

plt.title('Daily=false, Weekly=true, Yearly=false')
plt.hist(results_pd,bins=10)
plt.xticks(np.arange(0.0,1.0, 0.1))
plt.axvline(results_pd.mean()[0], color='r', linestyle='dashed', linewidth=4)
plt.axvline(results_pd.median()[0], color='g', linestyle='dashed', linewidth=4)
results_std = results_pd.std()
plt.axvline(results_pd.median()[0] + results_std[0], color='y', linestyle='dashed', linewidth=4)
plt.axvline(results_pd.median()[0] - results_std[0], color='y', linestyle='dashed', linewidth=4)

red_patch = mpatches.Patch(color='red', label='Mean')
green_patch = mpatches.Patch(color='green', label='Median')
yellow_patch = mpatches.Patch(color='yellow', label='One Standard Deviation')


plt.legend(handles=[red_patch,green_patch,yellow_patch],prop={'size':25})



plt.savefig("FIGS/all289.png")
plt.show()
```

```python
def unwrap(lst):
    ret = []
    for inLst in lst:
        ret.append(inLst[0])
    return ret

def fromSeriesToDF(df,series,IDX):
    series = df.iloc[IDX,:]
    dfSeries = series.to_frame()
    newdata = { 'ds':dfSeries.axes[0][1:].tolist(),'y':unwrap(dfSeries.values.tolist()[1:]) }
    newframe = pd.DataFrame(data=newdata)
    return newframe

def getTrainTest(df,sz,prc):
    proc = 100 - prc
    until = (proc * sz) // 100
    return df.iloc[:until],df.iloc[until:],until


def doForSeries(df,series,IDX,with_daily_seasonality = False,with_weekly_seasonality = False,with_yearly_seasonality = False):
    print("I AM HERE")
    newdf = fromSeriesToDF(df,series,IDX).dropna()
    df_size = newdf.shape[0]
    train_df,test_df,until = getTrainTest(newdf,df_size,30)

    
    my_model = Prophet(interval_width=0.95,daily_seasonality=with_daily_seasonality,weekly_seasonality=with_weekly_seasonality,yearly_seasonality=with_yearly_seasonality)
    my_model.fit(train_df)
    future_dates = my_model.make_future_dataframe(periods=df_size-until)
    forecast = my_model.predict(future_dates)
    
    forecasted_values = unwrap(forecast[['yhat']].values.tolist())
    test_values = unwrap(test_df[['y']].values.tolist())
    return test_values,forecasted_values,future_dates,until
```

```python
IDX = 1001
row = df.iloc[IDX]
test_values,forecasted_values,future_dates,until = doForSeries(df,df.iloc[IDX],IDX,True,True,True)

plt.figure(figsize=(20,10))
plt.plot(future_dates,row[1:].values.tolist())
plt.plot(future_dates,forecasted_values,'r',linewidth=2)

red_patch = mpatches.Patch(color='red', label='Forecasted')
blue_patch = mpatches.Patch(color='blue', label='Original')
yellow_patch = mpatches.Patch(color='green', label='Test from this point')


plt.axvline(future_dates.iloc[until][0], color='g', linestyle='dashed', linewidth=4)


plt.legend(handles=[red_patch,blue_patch,yellow_patch])

plt.savefig("FIGS/series" + str(IDX) +".png")
plt.show()
```

```python
def danyParser(filename):
    ret = []
    with open(filename,"r") as f:
        for line in f:
            ret = list(map(float,line.split(",")))
    return ret
```

```python
deep1 = danyParser("DANI/deep.out")

deep_pd = pd.DataFrame(deep1)
```

```python
plt.figure(figsize=(20,10))

plt.hist(deep_pd,bins=10)
plt.xticks(np.arange(0.0,1.0, 0.1))
plt.axvline(deep_pd.mean()[0], color='r', linestyle='dashed', linewidth=4)
plt.axvline(deep_pd.median()[0], color='g', linestyle='dashed', linewidth=4)
deep_std = deep_pd.std()
plt.axvline(deep_pd.median()[0] + deep_std[0], color='y', linestyle='dashed', linewidth=4)
plt.axvline(deep_pd.median()[0] - deep_std[0], color='y', linestyle='dashed', linewidth=4)

red_patch = mpatches.Patch(color='red', label='Mean')
green_patch = mpatches.Patch(color='green', label='Median')
yellow_patch = mpatches.Patch(color='yellow', label='One Standard Deviation')


plt.legend(handles=[red_patch,green_patch,yellow_patch],prop={'size':25})

plt.savefig("FIGS/dani.png")
plt.show()
```
