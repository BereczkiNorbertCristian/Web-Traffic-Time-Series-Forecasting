```python
lst = []
for i in range(0,100):
    row = df.loc[i][1:]
    lst.append(row.mean())

series = pd.Series(lst)
plt.subplot(1,2,1)
series.plot(kind='hist',figsize=(15,5))
plt.subplot(1,2,2)
series.plot(kind='kde')
plt.show()
```

```python
plt.subplot(1,2,1)
series.plot(kind='pie',figsize=(15,5))
plt.subplot(1,2,2)
series.plot(kind='box')
plt.show()
```

```python
lst = []
lststd = []
for i in range(0,100):
    row = df.loc[i][1:]
    lst.append(row.median())
    lststd.append(row.std())

series = pd.Series(lst)
plt.subplot(1,2,1)
series.plot(kind='hist',figsize=(15,5))
plt.subplot(1,2,2)
series.plot(kind='kde')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.scatter(lst,lststd)
plt.show()
```

```python
import scipy

pvl = scipy.stats.pearsonr(lst,lststd)
pvl2 = np.corrcoef(lst, lststd)[0, 1]
# pvl
pvl2
```
