```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def readFile(filename):
    lst = []
    with open(filename,"r") as f:
        for line in f:
            x = float(line.rstrip("\n"))
            if x > 10: continue
            lst.append(x)
    return lst
```

```python
out_root = "OUT/"
lstTFT = readFile(out_root + "TFT.out")
lstTFF = readFile(out_root + "TFF.out")
lstFFF = readFile(out_root + "FFF.out")
lstFFT = readFile(out_root + "FFT.out")
lstFTF = readFile(out_root + "FTF.out")
```

```python
tft_pd = pd.DataFrame(lstTFT)
tff_pd = pd.DataFrame(lstTFF)
fff_pd = pd.DataFrame(lstFFF)
fft_pd = pd.DataFrame(lstFFT)
ftf_pd = pd.DataFrame(lstFTF)
```

```python
plt.figure(figsize=(15,20))

plt.subplot(5,1,1)
plt.title('Daily=true, Weekly=false, Yearly=true')
plt.hist(lstTFT,bins=10)
plt.axvline(tft_pd.mean()[0], color='r', linestyle='dashed', linewidth=2)
plt.axvline(tft_pd.median()[0], color='g', linestyle='dashed', linewidth=2)

plt.subplot(5,1,2)
plt.title('Daily=true, Weekly=false, Yearly=false')
plt.hist(lstTFF,bins=10)
plt.axvline(tff_pd.mean()[0], color='r', linestyle='dashed', linewidth=2)
plt.axvline(tff_pd.median()[0], color='g', linestyle='dashed', linewidth=2)

plt.subplot(5,1,3)
plt.title('Daily=false, Weekly=false, Yearly=false')
plt.hist(lstFFF,bins=10)
plt.axvline(fff_pd.mean()[0], color='r', linestyle='dashed', linewidth=2)
plt.axvline(fff_pd.median()[0], color='g', linestyle='dashed', linewidth=2)

plt.subplot(5,1,4)
plt.title('Daily=false, Weekly=false, Yearly=true')
plt.hist(lstFFT,bins=10)
plt.axvline(fft_pd.mean()[0], color='r', linestyle='dashed', linewidth=2)
plt.axvline(fft_pd.median()[0], color='g', linestyle='dashed', linewidth=2)

plt.subplot(5,1,5)
plt.title('Daily=false, Weekly=true, Yearly=false')
plt.hist(lstFTF,bins=10)
plt.axvline(ftf_pd.mean()[0], color='r', linestyle='dashed', linewidth=2)
plt.axvline(ftf_pd.median()[0], color='g', linestyle='dashed', linewidth=2)

plt.show()
```
